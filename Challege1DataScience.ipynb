### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



## Facturacion total de cada tienda

```python
#Calculando
ingreso_tienda1 = tienda['Precio'].sum()
ingreso_tienda2 = tienda2['Precio'].sum()
ingreso_tienda3 = tienda3['Precio'].sum()
ingreso_tienda4 = tienda4['Precio'].sum()
```

```python

ingresos_totales = pd.DataFrame({
    'Tienda': ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
    'Ingreso Total': [
        ingreso_tienda1,
        ingreso_tienda2,
        ingreso_tienda3,
        ingreso_tienda4
    ]
})

ingresos_totales.style.format({'Ingreso Total': '{:,.2f}'})
```

```python
ingresos_totales = ingresos_totales.sort_values(
    by='Ingreso Total',
    ascending=False
)
```

```python
from enum import auto
#Grafico pastel del ingreso total de las 4 tiendas
#set_index('Tienda') → las tiendas se convierten en etiquetas
#['Ingreso Total'] → valores del gráfico
#plot.pie() → gráfico de pastel desde Pandas
#autopct → muestra porcentajes
#startangle → mejor orientación visual

ingresos_totales.set_index('Tienda')['Ingreso Total'].plot.pie(
    figsize=(6,6),
    autopct='%1.1f%%',
    startangle=90,
    title='Ingresos total de cada tienda'
)

```

# 2. Ventas por categoría

```python
#Que categoria de productos se vende mas por cada tienda
#groupby() → agrupar por categoría
#count() o size() → contar ventas
#reset_index() → convertir resultado en tabla
#sort_values() → ordenar por popularidad
```

```python
#Agrupar por categoria y contar las filas
#Formula base tienda.groupby('Categoría del Producto').size()
ventas_categoria_tienda = (
    tienda
    .groupby('Categoría del Producto')
    .size()
    .reset_index(name='Cantidad Vendida')
    .sort_values(by='Cantidad Vendida', ascending=True)
)

ventas_categoria_tienda2 = (
    tienda2
    .groupby('Categoría del Producto')
    .size()
    .reset_index(name='Cantidad Vendida')
    .sort_values(by='Cantidad Vendida', ascending=True)
)

ventas_categoria_tienda3 = (
    tienda3
    .groupby('Categoría del Producto')
    .size()
    .reset_index(name='Cantidad Vendida')
    .sort_values(by='Cantidad Vendida', ascending=True)
)

ventas_categoria_tienda4 = (
    tienda4
    .groupby('Categoría del Producto')
    .size()
    .reset_index(name='Cantidad Vendida')
    .sort_values(by='Cantidad Vendida', ascending=True)
)
```

```python
#Grafico de las categorias mas vendidas de la tienda 1
aTienda_Grafico = ventas_categoria_tienda.set_index('Categoría del Producto')['Cantidad Vendida'].plot(kind='barh',figsize=(12,8), color='purple')
aTienda_Grafico.set_title('Cantidad de productos vendidos por categoría - Tienda 1')
aTienda_Grafico.set_xlabel('Cantidad Vendida')
aTienda_Grafico.set_ylabel('Categoría del Producto')

# Agregar etiquetas de valor
for i, v in enumerate(ventas_categoria_tienda['Cantidad Vendida']):
    aTienda_Grafico.text(v, i, str(v), va='center')
```

## Ingreso total por categoria de cada tienda


```python
ingreso_categoria_tienda = (
    tienda.groupby('Categoría del Producto')['Precio']
    .sum()
    .reset_index()
    .rename(columns={'Precio': 'Ingreso por Categoría Tienda'})
    .sort_values(by='Ingreso por Categoría Tienda', ascending=False)

)

ingreso_categoria_tienda2 = (
    tienda2.groupby('Categoría del Producto')['Precio']
    .sum()
    .reset_index()
    .rename(columns={'Precio': 'Ingreso por Categoría Tienda'})
    .sort_values(by='Ingreso por Categoría Tienda', ascending=False)

)

ingreso_categoria_tienda3 = (
    tienda3.groupby('Categoría del Producto')['Precio']
    .sum()
    .reset_index()
    .rename(columns={'Precio': 'Ingreso por Categoría Tienda'})
    .sort_values(by='Ingreso por Categoría Tienda', ascending=False)

)

ingreso_categoria_tienda4 = (
    tienda4.groupby('Categoría del Producto')['Precio']
    .sum()
    .reset_index()
    .rename(columns={'Precio': 'Ingreso por Categoría Tienda'})
    .sort_values(by='Ingreso por Categoría Tienda', ascending=False)

)
```
